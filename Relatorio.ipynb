# Análise Avaliação Livros



```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
books = pd.read_csv("BX-Books.csv", error_bad_lines=False,low_memory=False,sep=';')
ratings = pd.read_csv("BX-Book-Ratings.csv", error_bad_lines=False,low_memory=False,sep=';')
users = pd.read_csv("BX-Users.csv", error_bad_lines=False,low_memory=False,sep=';')
```

```python
users.tail(10)
```

```python
ratings.tail(10)
```

```python
df = pd.merge(users, ratings, on='User-ID')
```

```python
df.tail(10)
```

```python
books.tail(10)
```

```python
df = pd.merge(df, books, on='ISBN')
```

```python
df.tail(10)
```

```python
data = df[['User-ID','Age','Book-Rating','Book-Title','Book-Author','Year-Of-Publication','Publisher','ISBN']]
```

```python
data.tail(10)
```

```python
books = data['ISBN'].value_counts().reset_index()
books.columns = ['ISBN', 'Count-Book']
data = pd.merge(data, books)
```

```python
years = data['Year-Of-Publication'].value_counts().reset_index()
years.columns = ['Year-Of-Publication', 'Count-Year']
data = pd.merge(data, years)
```

```python
user = data['User-ID'].value_counts().reset_index()
user.columns = ['User-ID', 'Count-User']
data = pd.merge(data, user)
```

```python
data['Book-Title'] = data['Book-Title'].str.capitalize()
data['Book-Author'] = data['Book-Author'].str.capitalize()
```

```python
plt.figure(figsize=(8,8))
plt.hist(data['Age'],bins=[0,20,40,60,80,100],histtype ='bar',rwidth=0.95)
plt.grid(True)
plt.ylabel('Frequencia')
plt.xlabel('Idades')
plt.show()
```

```python
print(data['Age'].mean())
print(data['Age'].std())
print(data['Age'].quantile(0.5))
```

```python
print(data[(data['Age']<100) & (data['Age']>0)]['Age'].mean())
print(data[(data['Age']<100) & (data['Age']>0)]['Age'].std())
print(data[(data['Age']<100) & (data['Age']>0)]['Age'].quantile(0.5))
```

```python
print(len(data[(data['Age']>100) | (data['Age']==0)]))
print(len(data[(data['Age']<100) & (data['Age']>0)]))
```

```python
media_rating = data[(data['Count-User']>50)]
media_rating = media_rating.groupby('Age')['Book-Rating'].mean().reset_index()
plt.figure(figsize=(15,10))
plt.scatter(media_rating['Age'],media_rating['Book-Rating'])
plt.grid(True)
plt.ylabel('Avaliação')
plt.xlabel('Idade')
plt.title("Média de Avaliações por idade.") 
plt.show()
```

```python
media_20_60 = data[(data['Count-User']>50) & (data['Age']>20) & (data['Age']<60)]['Book-Rating']
media_20_60.mean()
```

```python
year_cited = data['Year-Of-Publication'].value_counts().reset_index()
year_cited.columns = ['Year-Of-Publication', 'Count-Year']
year_cited = year_cited.sort_values('Count-Year')[-50:]
plt.figure(figsize=(8,10))
plt.barh(year_cited['Year-Of-Publication'],year_cited['Count-Year'])
plt.xlabel("Número de Citações") 
plt.ylabel("Ano") 
plt.title("Anos Mais Citados") 
plt.grid(True)
plt.show()
```

```python
media_year = data[(data['Count-Year']>1000) & (data['Year-Of-Publication']!='0')]
media_year = media_year.groupby('Year-Of-Publication')['Book-Rating'].mean().reset_index().sort_values('Book-Rating')[-50:]
plt.figure(figsize=(8,10))
plt.barh(media_year['Year-Of-Publication'],media_year['Book-Rating'])
plt.xlabel("Pontuação Média") 
plt.ylabel("Ano") 
plt.title("Média dos Anos") 
plt.grid(True)
plt.show()
```

```python
media_book = data[(data['Count-Book']>100) & (data['Year-Of-Publication']!='0')]
media_book = media_book.groupby('Book-Title')['Book-Rating'].mean().reset_index().sort_values('Book-Rating')[-50:]
plt.figure(figsize=(8,10))
plt.barh(media_book['Book-Title'],media_book['Book-Rating'])
plt.xlabel("Pontuação Média") 
plt.ylabel("Livro") 
plt.title("50 melhores livros (Minimo de 100 Avaliações)") 
plt.grid(True)
plt.show() 
```

```python
media_author = data[(data['Count-Book']>100) & (data['Year-Of-Publication']!='0')]
media_author = media_author.groupby('Book-Author')['Book-Rating'].mean().reset_index().sort_values('Book-Rating')[-50:]
plt.figure(figsize=(8,10))
plt.barh(media_author['Book-Author'],media_author['Book-Rating'])
plt.xlabel("Pontuação Média") 
plt.ylabel("Autor") 
plt.title("Média dos 50 Melhores Autores") 
plt.grid(True)
plt.show()
```

```python
media_publisher = data[(data['Count-Book']>100) & (data['Year-Of-Publication']!='0')]
media_publisher = media_publisher.groupby('Publisher')['Book-Rating'].mean().reset_index().sort_values('Book-Rating')[-50:]
plt.figure(figsize=(8,10))
plt.barh(media_publisher['Publisher'],media_publisher['Book-Rating'])
plt.xlabel("Pontuação Média") 
plt.ylabel("Editora") 
plt.title("Média das 50 Melhores Editoras") 
plt.grid(True)
plt.show()
```

```python
author_cited = data[(data['Count-Book']>100) & (data['Year-Of-Publication']!='0')]
author_cited = author_cited['Book-Author'].value_counts().reset_index()
author_cited.columns = ['Book-Author', 'Count-Author']
author_cited = author_cited.sort_values('Count-Author')[-50:]
plt.figure(figsize=(8,10))
plt.barh(author_cited['Book-Author'],author_cited['Count-Author'])
plt.xlabel("Número de Citações") 
plt.ylabel("Autor") 
plt.title("50 Autores Mais Citados") 
plt.grid(True)
plt.show()
```

```python
books_cited = data[(data['Count-Book']>100) & (data['Year-Of-Publication']!='0')]
books_cited = books_cited['Book-Title'].value_counts().reset_index()
books_cited.columns = ['Book-Title', 'Count-Book']
books_cited = books_cited.sort_values('Count-Book')[-50:]
plt.figure(figsize=(8,10))
plt.barh(books_cited['Book-Title'],books_cited['Count-Book'])
plt.xlabel("Número de Citações") 
plt.ylabel("Livro") 
plt.title("50 Livros Mais Citados") 
plt.grid(True)
plt.show()
```
